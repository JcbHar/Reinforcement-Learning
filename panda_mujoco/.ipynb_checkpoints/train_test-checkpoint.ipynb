<h1 style="color: orange; font-family: 'Roboto', sans-serif;">Setup</h1>

```python
import os
import numpy as np

import mujoco
import mujoco.viewer

from agents.sac_agent import SACAgent

from wrapper.panda_env import PandaEnv

from wrapper.methods import print_mujoco_info
from wrapper.methods import plot_monitor_rewards
from wrapper.methods import force_print_to_terminal

from stable_baselines3.common.monitor import Monitor

from stable_baselines3.common.env_util import make_vec_env
```

```python
log_dir = "logs"
os.makedirs(log_dir, exist_ok=True)

mjcf_path = "models/environment/scene.xml"

agent_max_steps = 600_000
env_max_steps = 2_000

env = Monitor(PandaEnv(mjcf_path, max_steps=env_max_steps), log_dir)

agent = SACAgent(env)
```

```python
force_print_to_terminal()
```

<div style="height: 60px;"></div>

---

<h1 style="color: orange; font-family: 'Roboto', sans-serif;">Train</h1>

```python
agent = SACAgent(env)

agent.train(timesteps=agent_max_steps)

plot_monitor_rewards()
```

<div style="height: 60px;"></div>

---

<h1 style="color: orange; font-family: 'Roboto', sans-serif;">Test</h1>

```python
episode_count = 0
max_episodes = agent_max_steps // env_max_steps

obs, _ = env.reset()
done = False
truncated = False

env.unwrapped.launch_viewer()  

while episode_count < max_episodes:
    action, _ = agent.predict(obs)
    obs, reward, done, truncated, _ = env.step(action)
    env.render()

    if done or truncated:
        episode_count += 1
        print(f"episode {episode_count} done")

        if episode_count >= max_episodes:
            print("simulation complete")
            break

        obs, _ = env.reset()
        done = False
        truncated = False
```
